```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data={
    "age" : [20,30,40,50,60,70,80],
    "salary":[2000,3000,4000,6000,3000,5000,9000]}
df=pd.DataFrame(data)
sns.barplot(x="age",y="salary",data=df)
plt.show()
```

```python
sns.scatterplot(x="age",y="salary",data=df,color="black")
plt.title("salary vs age ploting")
plt.show()
```

```python
sns.lineplot(x="age",y="salary",data=df,linewidth=3)
plt.show()
```

```python
data={
    "position": ["HR","Finance",
    "IT",
    "Sales",
    "Marketing",
    "Operations",
    "Support"],
    "salary":[2000,3000,4000,6000,3000,5000,9000]}

df=pd.DataFrame(data)
sns.barplot(x="position",y="salary",data=df)
plt.show()

    
```

```python
sns.boxplot(y="salary",data=df)
plt.show()
```

```python
sns.violinplot(x="position",y="salary",data=df)
plt.title("violinplot")
plt.show()
```
