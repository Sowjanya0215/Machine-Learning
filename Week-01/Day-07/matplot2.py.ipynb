```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data={"salaries": [10000,30000,25000, 30000, 28000, 35000, 40000]}
df=pd.DataFrame(data)
```

```python
plt.hist(df["salaries"])
plt.show()
```

```python
plt.boxplot(df["salaries"])
plt.xlabel("num")
plt.show()
```

```python
df["dept"]=["HR", "IT", "Finance", "IT", "HR", "Marketing", "Finance"]
df.head()
```

```python
df.info()
```

```python
count=df["dept"].value_counts()
count
```

```python
plt.pie(count,labels=count.index,autopct="%1.1f")
plt.show()
```

```python
plt.bar(count.index,count,color=["green","pink","red","blue"])
```

```python
df["Age"]=[30,40,56,60,78,80,90]
df.head()
```

```python
plt.scatter(df["Age"],df["salaries"],color="red")
plt.xlabel("Ages")
plt.ylabel("salaries")
plt.title("Employee status")
plt.show()
```

```python
plt.plot(df["Age"],df["salaries"],color="red",marker="o",linestyle=":",linewidth="2")
plt.xlabel("Ages")
plt.ylabel("salaries")
plt.title("Employee status")
plt.show()

```

```python
sort_age=df.sort_values(by="Age")
```

```python
sort_age["Age"]
```

```python
plt.bar(sort_age["Age"],df["salaries"],color="pink",linewidth=2,width=0.8)

plt.xlabel("Ages")
plt.ylabel("salaries")
plt.title("Employee status")
plt.show()
```

```python
subjects = ["Math","Science","English"]
marks = [85,90,78]

plt.bar(subjects, marks, color="skyblue")
plt.title("Student Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
subjects = ["Math","Science","English"]
marks = [85,90,78]

plt.barh(marks,subjects)
plt.xlabel("subjects")
plt.ylabel("marks")
plt.show()
```

```python
ages = [20,21,22,23,22,21,20,24,25,22]

plt.hist(ages, bins=5)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
